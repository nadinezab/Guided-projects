## Decomposition

Decomposition is a tool used for time series analysis. Provides a structured way to think about a time series forecasting problem.

The statsmodels librariy has a `seasonal_decompose()` method to automatically decompose a time series. It requires specifying if the model is additive or multiplicative.

Before looking at the real world dataset, let us explore basic examples.

### Simulated examples

```python
from random import randrange
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

**Additive decomposition**

```python
series = [i+randrange(10) for i in range(1,100)]
result = seasonal_decompose(series, model='additive', freq=1)
result.plot()
plt.show()
```

**Multiplicative decomposition**

```python
series = [i**2.0 for i in range(1,100)]
result = seasonal_decompose(series, model='multiplicative', freq=1)
result.plot()
plt.show()
```

### Dataset

```python
series = pd.read_csv('datasets/airline/airline-passengers.csv')
series.head()
```

```python
# Change the data type of the 'Month' column
series['Month'] = pd.to_datetime(series['Month'])
```

```python
# Set the 'Month' column as the index
series = series.set_index('Month')
```

```python
series.plot(figsize=(10,4), color='blue')
plt.show()
```

We will assume a multiplicative model.

```python
result = seasonal_decompose(series, model='multiplicative')
result.plot()
plt.show()
```
